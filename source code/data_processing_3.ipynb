```python
import numpy as np
import pandas as pd

import warnings

warnings.filterwarnings('ignore')

Calcium = pd.read_csv('cal_result.csv')
Creatinine = pd.read_csv('cre_result.csv')
Glucose= pd.read_csv('glu_result.csv')
pH=pd.read_csv('ph_result.csv')
Urea_Nitrogen= pd.read_csv('urea_result.csv')
Urine=pd.read_csv('urine_result.csv')
patients_informations=pd.read_csv('patients_informations.csv')
```

```python
Blood_Calcium=Calcium.drop(['Unnamed: 0'], axis=1)
Blood_Creatinine=Creatinine.drop(['Unnamed: 0'], axis=1)
Blood_Glucose=Glucose.drop(['Unnamed: 0'], axis=1)
Blood_pH=pH.drop(['Unnamed: 0'], axis=1)
Blood_Urea_Nitrogen=Urea_Nitrogen.drop(['Unnamed: 0'], axis=1)
Urine=Urine.drop(['Unnamed: 0'], axis=1)
```

```python
# Add column names to each of the six tables generated in data_processing_2
Blood_Calcium.columns = ['SUBJECT_ID','HADM_ID','SEQ_NUM','Blood_Calcium']
Blood_Creatinine.columns = ['SUBJECT_ID','HADM_ID','SEQ_NUM','Blood_Creatinine']
Blood_Glucose.columns = ['SUBJECT_ID','HADM_ID','SEQ_NUM','Blood_Glucose']
Blood_pH.columns = ['SUBJECT_ID','HADM_ID','SEQ_NUM','Blood_pH']
Blood_Urea_Nitrogen.columns = ['SUBJECT_ID','HADM_ID','SEQ_NUM','Blood_Urea_Nitrogen']
Urine.columns = ['SUBJECT_ID','HADM_ID','SEQ_NUM','Urine']
```

```python
All_result_1=pd.merge(Blood_Calcium,Blood_Creatinine,on=['SUBJECT_ID', 'HADM_ID','SEQ_NUM'],how='left')
All_result_1
```

```python
All_result_2=pd.merge(All_result_1,Blood_Glucose,on=['SUBJECT_ID', 'HADM_ID','SEQ_NUM'],how='left')
All_result_2
```

```python
All_result_3=pd.merge(All_result_2,Blood_pH,on=['SUBJECT_ID', 'HADM_ID','SEQ_NUM'],how='left')
All_result_3
```

```python
All_result_4=pd.merge(All_result_3,Blood_Urea_Nitrogen,on=['SUBJECT_ID', 'HADM_ID','SEQ_NUM'],how='left')
All_result_4
```

```python
All_result=pd.merge(All_result_4,Urine,on=['SUBJECT_ID', 'HADM_ID','SEQ_NUM'],how='left')
All_result
```

```python
#Merge the liquid test results table, All_result, with the generated patients_informations in data_processing_1
result_test_1=pd.merge(patients_informations,All_result,on=['SUBJECT_ID', 'HADM_ID','SEQ_NUM'])
result_test_1
```

```python
result_test_2 = result_test_1.drop(['Unnamed: 0', 'ICD9_CODE'],axis=1)
result_test_2
```

```python
result_test_3 = result_test_2.drop(['SUBJECT_ID','HADM_ID', 'SEQ_NUM'],axis=1)
```

```python
result_test_3['PATIENT_ID'] = result_test_3.index
result_test_3
```

```python
columns = list(result_test_3)
# move the column to head of list using index, pop and insert
columns.insert(0, columns.pop(columns.index('PATIENT_ID')))
columns.insert(14, columns.pop(columns.index('Death_Flag')))
columns.insert(13, columns.pop(columns.index('hypertension')))
columns.insert(12, columns.pop(columns.index('diabetes')))
# use loc to reorder
result_test_3 = result_test_3.loc[:, columns]
result_test_3
```

```python
#Convert data types to int
result_test_3[['diabetes']] = result_test_3[['diabetes']].astype(int)
result_test_3[['hypertension']] = result_test_3[['hypertension']].astype(int)
result_test_3
```

```python
#Filter for non-null values
result_test_3 = result_test_3[result_test_3['RELIGION'].notna()]
result_test_3 = result_test_3[result_test_3['MARITAL_STATUS'].notna()]
result_test_3 = result_test_3[result_test_3['ETHNICITY'].notna()]
result_test_3
```

```python
#Removal of properties with only a few values
result_test_4=result_test_3.drop(result_test_3[result_test_3['ETHNICITY']=='PATIENT DECLINED TO ANSWER'].index)
result_test_5=result_test_4.drop(result_test_4[result_test_4['ETHNICITY']=='UNABLE TO OBTAIN'].index)
result_test_6=result_test_5.drop(result_test_5[result_test_5['ETHNICITY']=='UNKNOWN/NOT SPECIFIED'].index)
result_test_7=result_test_6.drop(result_test_6[result_test_6['MARITAL_STATUS']=='UNKNOWN (DEFAULT)'].index)
result_test_8=result_test_7.drop(['RELIGION'], axis=1)
result_test_8
```

```python
White = result_test_8.loc[result_test_8['ETHNICITY'].str.contains('WHITE')]
Black = result_test_8.loc[result_test_8['ETHNICITY'].str.contains('BLACK')]
Asian = result_test_8.loc[result_test_8['ETHNICITY'].str.contains('ASIAN')]
Hispanic = result_test_8.loc[result_test_8['ETHNICITY'].str.contains('HISPANIC')]
```

```python
White['ETHNICITY'] = 'WHITE'
Black['ETHNICITY'] = 'BLACK'
Asian['ETHNICITY'] = 'ASIAN'
Hispanic['ETHNICITY'] = 'HISPANIC'
```

```python
final_result = pd.concat([White,Black,Asian,Hispanic])
final_result.sort_values(by=['PATIENT_ID'], ascending=True, inplace=True)
final_result
```

```python
# move the column to head of list using index, pop and insert
final_result.reset_index(inplace=True)
final_data = final_result.drop(['index', 'PATIENT_ID'],axis=1)
final_data['PATIENT_ID'] = final_data.index
columns = list(final_data)
columns.insert(0, columns.pop(columns.index('PATIENT_ID')))
final_data = final_data.loc[:, columns]
final_data
```

```python
final_data.to_csv('final_data.csv')
```
