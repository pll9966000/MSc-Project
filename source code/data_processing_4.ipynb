```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
final_data = pd.read_csv('final_data.csv')
```

```python
Renal_Failure_data=final_data.drop(['Unnamed: 0'], axis=1)
```

```python
#The individual values in the string column are represented by numbers.
Renal_Failure_data['GENDER']=Renal_Failure_data['GENDER'].map({'M':1,'F':2}).astype(int)
Renal_Failure_data['MARITAL_STATUS']=Renal_Failure_data['MARITAL_STATUS'].map({'DIVORCED':1,'LIFE PARTNER':2,'MARRIED':3,'SEPARATED':4,'SINGLE':5,'WIDOWED':6}).astype(int)
Renal_Failure_data['ETHNICITY']=Renal_Failure_data['ETHNICITY'].map({'ASIAN':1,'BLACK':2,'HISPANIC':3,'WHITE':4}).astype(int)
Renal_Failure_data
```

```python
#Check the table for null values
Renal_Failure_data.info()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

pipeline = Pipeline([
    # Deal with the NaN value
    ('imputer',SimpleImputer(strategy="median"))
])
```

```python
features_prepared = pipeline.fit_transform(Renal_Failure_data)
```

```python
final_data_1=pd.DataFrame(features_prepared)
final_data_1
```

```python
final_data_1.columns=[Renal_Failure_data.columns.values]
final_data_1
```

```python
final_data_1[['PATIENT_ID']] = final_data_1[['PATIENT_ID']].astype(int)
final_data_1[['AGE']] = final_data_1[['AGE']].astype(int)
final_data_1[['GENDER']] = final_data_1[['GENDER']].astype(int)
final_data_1[['MARITAL_STATUS']] = final_data_1[['MARITAL_STATUS']].astype(int)
final_data_1[['ETHNICITY']] = final_data_1[['ETHNICITY']].astype(int)
final_data_1[['diabetes']] = final_data_1[['diabetes']].astype(int)
final_data_1[['hypertension']] = final_data_1[['hypertension']].astype(int)
final_data_1[['Death_Flag']] = final_data_1[['Death_Flag']].astype(int)
final_data_1
```

```python
#Observe the balance of data
final_data_1.Death_Flag.value_counts()
```

```python
corrRF = final_data_1.corr()
corrRF
```
