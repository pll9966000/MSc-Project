```python
import numpy as np
import pandas as pd
# Read the tables needed for the experiment.
labevents = pd.read_csv('LABEVENTS.csv')
labitems = pd.read_csv('D_LABITEMS.csv')
patients_informations=pd.read_csv('patients_informations.csv')
```

```python
labevents_1 = labevents.drop(['ROW_ID', 'CHARTTIME'], axis=1) 
labevents_1
```

```python
labitems
```

```python
labitems_1 = labitems.drop(['ROW_ID', 'CATEGORY','LOINC_CODE'], axis=1) 
labitems_1
```

```python
#Combine the patients_informations.csv and LABEVENTS.csv generated by data_processing_1.
labevents_items=pd.merge(labevents_1,labitems_1,on=['ITEMID'])
labevents_items
```

```python
patients_informations_1 = patients_informations.drop(['Unnamed: 0'], axis=1) 
patients_informations_1
```

```python
patients_labevents_items=pd.merge(patients_informations_1,labevents_items,on=['SUBJECT_ID','HADM_ID'])
patients_labevents_items
```

```python
patients_labevents_items_1=patients_labevents_items.drop(['ICD9_CODE'],axis=1)
patients_labevents_items_1
```

```python
patients_labevents_items_pH=patients_labevents_items_1[(patients_labevents_items_1.FLUID=='Blood') & (patients_labevents_items_1.LABEL=='pH')]
patients_labevents_items_pH
```

```python
patients_labevents_items_pH_1=patients_labevents_items_pH.values
patients_labevents_items_pH_1
```

```python
name1=patients_labevents_items_pH[['SUBJECT_ID']]
name2=name1.drop_duplicates()
name3 = name2['SUBJECT_ID'].tolist()
name3
```

```python
#Calculating the proportion of abnormal pH in the blood
i=0
l=[]
ph = [([] * 3) for i in range(len(name3)-1)]
for i in range(len(name3)-1):
    s=0 #The total number of pH tests performed on a single blood test for a particular patient
    a=0
    for j in range(102852):
        # Compare two adjacent data for a patient to see if they are the same
        if name3[i]==patients_labevents_items_pH_1[j,0] and patients_labevents_items_pH_1[j,0]== patients_labevents_items_pH_1[j+1,0] and patients_labevents_items_pH_1[j,1]== patients_labevents_items_pH_1[j+1,1] and patients_labevents_items_pH_1[j,2]== patients_labevents_items_pH_1[j+1,2]:
            s=s+1
            ph[i]=[patients_labevents_items_pH_1[j,0],patients_labevents_items_pH_1[j,1],patients_labevents_items_pH_1[j,2]]
            if patients_labevents_items_pH_1[j,13]<7.35:# An abnormal value is determined when pH < 7.35
                a=a+1
        else:
            j=j-1
    l.append(a/(s+1)) #a/(s+1) is the formula for the probability of outliers
    
print(len(l))
print(len(ph))
print(l)
print(ph)
        
```

```python
pd_l = pd.DataFrame(l)
pd_ph=pd.DataFrame(ph)
ph_result = pd.concat([pd_ph, pd_l], axis=1)
ph_result_1 = ph_result.dropna(axis=0, thresh=3)
ph_result_1
```

```python
ph_result.to_csv('ph_result.csv')
```
