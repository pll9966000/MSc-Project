```python
import numpy as np
import pandas as pd
import warnings

warnings.filterwarnings('ignore')
```

```python
# Read the tables needed for the experiment.
admissions = pd.read_csv('ADMISSIONS.csv')
patients = pd.read_csv('PATIENTS.csv')
diagnoses = pd.read_csv('DIAGNOSES_ICD.csv')
d_items = pd.read_csv('D_ITEMS.csv')
```

```python
patients.head()
```

```python
patients_1 = patients.drop(['ROW_ID', 'DOD', 'DOD_HOSP', 'DOD_SSN', 'EXPIRE_FLAG'], axis=1) 
patients_1
```

```python
admissions.head()
```

```python
admissions_1 = admissions.drop(['ROW_ID', 'DIAGNOSIS', 'ADMISSION_TYPE', 'ADMISSION_LOCATION', 'DISCHARGE_LOCATION', 'INSURANCE','LANGUAGE', 'EDREGTIME', 'EDOUTTIME','HAS_CHARTEVENTS_DATA'], axis = 1)
admissions_1
```

```python
patients_admissions=pd.merge(patients_1,admissions_1,on=['SUBJECT_ID'])
patients_admissions.head()
```

```python
columns = list(patients_admissions)
# move the column to head of list using index, pop and insert
columns.insert(3, columns.pop(columns.index('DOB')))
# use loc to reorder
patients_admissions = patients_admissions.loc[:, columns]
patients_admissions.head()
```

```python
diagnoses.head()
```

```python
# Find out about patients with acute renal failure
diagnoses_1 = diagnoses.drop(['ROW_ID'], axis=1)
diagnoses_2 = diagnoses_1.loc[diagnoses_1['ICD9_CODE']=='5849']
diagnoses_2
```

```python
patients_diagnoses=pd.merge(diagnoses_2,patients_admissions,on=['SUBJECT_ID', 'HADM_ID'])
patients_diagnoses.head()
```

```python
# SUBJECT_ID for patients with acute renal failure.
patients_ssj=patients_diagnoses[['SUBJECT_ID']]
patients_ssj_1=patients_ssj.drop_duplicates()
patients_ssj_1
```

```python
patients_ssj_diagnoses=pd.merge(patients_ssj_1,diagnoses_1,on=['SUBJECT_ID'])
patients_ssj_diagnoses
```

```python
#Identify patients with renal failure who also have diabetes
patients_ssj_diagnoses_1 = patients_ssj_diagnoses[patients_ssj_diagnoses.ICD9_CODE.isin(['2535','3572','5881','64800','64801','64802','64803','64804','7751','V771','V180','V1221'])]
patients_ssj_diagnoses_1
```

```python
# ID in patients with both acute renal failure and diabetes mellitus
patients_ssj_diagnoses_1_id=patients_ssj_diagnoses_1[['SUBJECT_ID']]
patients_ssj_diagnoses_1_id1=patients_ssj_diagnoses_1_id.drop_duplicates()
patients_ssj_diagnoses_1_id1
```

```python
#Identify patients with renal failure who also have hypertension
patients_ssj_diagnoses_2 = patients_ssj_diagnoses[patients_ssj_diagnoses.ICD9_CODE.isin(['36504','4010','4011','4019','40591','40599','4160','5723','V811','99791'])]
patients_ssj_diagnoses_2
```

```python
# ID in patients with both acute renal failure and hypertension
patients_ssj_diagnoses_2_id=patients_ssj_diagnoses_2[['SUBJECT_ID']]
patients_ssj_diagnoses_2_id2=patients_ssj_diagnoses_2_id.drop_duplicates()
patients_ssj_diagnoses_2_id2
```

```python
patients_ssj_diagnoses_1_id1.insert(patients_ssj_diagnoses_1_id1.shape[1], 'diabetes', 1)
```

```python
patients_ssj_diagnoses_1_id1
```

```python
patients_ssj_diagnoses_2_id2.insert(patients_ssj_diagnoses_2_id2.shape[1], 'hypertension', 1)
patients_ssj_diagnoses_2_id2
```

```python
patients_diagnoses_admissions_tnb = pd.merge(patients_diagnoses,patients_ssj_diagnoses_1_id1,how='left',on='SUBJECT_ID')
patients_diagnoses_admissions_tnb
```

```python
patients_diagnoses_admissions_gxy = pd.merge(patients_diagnoses_admissions_tnb,patients_ssj_diagnoses_2_id2,how='left',on='SUBJECT_ID')
patients_diagnoses_admissions_gxy
```

```python
patients_diagnoses_admissions_gxy['diabetes'].fillna(0, inplace=True)
```

```python
patients_diagnoses_admissions_gxy['hypertension'].fillna(0, inplace=True)
```

```python
patients_diagnoses_admissions_gxy
```

```python
patients_age = patients_diagnoses_admissions_gxy.drop( patients_diagnoses_admissions_gxy[patients_diagnoses_admissions_gxy['DOB'] < '2000-01-01'].index)
patients_age
```

```python
#Convert all the time types in the table to datetime.
patients_age['DOB'] = pd.to_datetime(patients_age['DOB'])
patients_age['ADMITTIME'] = pd.to_datetime(patients_age['ADMITTIME'])
patients_age['DISCHTIME'] = pd.to_datetime(patients_age['DISCHTIME'])
patients_age['DEATHTIME'] = pd.to_datetime(patients_age['DEATHTIME'])
```

```python
patients_age_1 = patients_age[(patients_age['DEATHTIME'].notnull()) & (patients_age['DEATHTIME'] != "")]
patients_age_1
```

```python
date1=patients_age_1['DOB']
date1
```

```python
# 'DEATHTIME' minus 'DOB' divided by 365 to obtain the patient's age
patients_age_1['AGE']=(patients_age_1['DEATHTIME']-patients_age_1['DOB'])/365
patients_age_1
```

```python
patients_age_1['AGE'] = patients_age_1['AGE'].dt.days
```

```python
patients_age_1
```

```python
patients_age_1.reset_index(inplace=True)
patients_death_1 = patients_age_1.drop(['index'],axis=1)
patients_death_1
```

```python
# Determining whether a patient with acute renal failure dies within 90 days
time=patients_death_1['DEATHTIME']-patients_death_1['ADMITTIME']
days=time.dt.days
days
a=[]
for i in range(1621):
    if days[i]<90:
        death=1 #If less than 90 days from admission to death, DEATH_Flag is assigned a value of 1
        a.append(death)
    else:
        death=0
        a.append(death)
patients_death_1['Death_Flag'] = pd.DataFrame(a)
patients_death_1
```

```python
patients_age_2 = patients_age[(patients_age['DEATHTIME'].isnull())]
patients_age_2
```

```python
#'ADMITTIME' minus 'DOB' divided by 365 to obtain the patient's age
patients_age_2['AGE']=(patients_age_2['ADMITTIME']-patients_age_2['DOB'])/365
patients_age_2['AGE'] = patients_age_2['AGE'].dt.days
patients_age_2
```

```python
patients_age_2['Death_Flag']=0
patients_age_2
```

```python
patients_death_2 = [patients_death_1, patients_age_2]
patients_death = pd.concat(patients_death_2)
patients_death
```

```python
columns = list(patients_death)
# move the column to head of list using index, pop and insert
columns.insert(3, columns.pop(columns.index('AGE')))
# use loc to reorder
patients_death = patients_death.loc[:, columns]
patients_death
```

```python
patients_death.reset_index(inplace=True)
patients_informations = patients_death.drop(['index'],axis=1)
patients_informations
```

```python
patients_informations=patients_informations.drop(['DOB', 'ADMITTIME', 'DISCHTIME', 'DEATHTIME', 'EXPIRE_FLAG', 'HOSPITAL_EXPIRE_FLAG'], axis=1) 
patients_informations
```

```python
patients_informations.to_csv('patients_informations.csv')
```
